```python
import pandas as pd
```

```python
df=pd.read_csv("temperatures.csv")
df
```

```python
Xd=df[["YEAR"]]
Yd=df[["ANNUAL"]]
print(Xd)
print(Yd)
```

```python
from sklearn.model_selection import train_test_split
Xd_train,Xd_test = train_test_split(Xd,train_size=0.60,random_state=1)
Yd_train,Yd_test = train_test_split(Yd,train_size=0.60,random_state=1)
print(Xd_test)
print(Yd_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(Xd_train,Yd_train)
```

```python
m=model.coef_
c=model.intercept_
print(m)
print(c)
```

```python
Y_predict=model.predict(Xd_test)
Y_predict
```

```python
import matplotlib.pyplot as plt
plt.scatter(Xd_test, Yd_test,color='red')
plt.plot(Xd_test, Y_predict,color='green')
```

```python
from sklearn import metrics
import numpy as np
print("MAE: ",metrics.mean_absolute_error(Yd_test,Y_predict))
print("MSE: ",metrics.mean_squared_error(Yd_test,Y_predict))
print("RMSE: ",np.sqrt(metrics.mean_squared_error(Yd_test,Y_predict)))
```
